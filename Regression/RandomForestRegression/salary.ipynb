## import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

## create a data frame

```python
df=pd.read_csv('salary.csv')
X=df.iloc[:,1:2].values
y=df.iloc[:,2:].values
```

## Training the model

```python
model=RandomForestRegressor(n_estimators=12,random_state=42)
model.fit(X,y)
```

```python
model.predict([[5.5]])
```

```python
X_grid=np.arange(X.min(),X.max(),0.01)
X_grid=X_grid.reshape(-1,1)
plt.scatter(X,y)
plt.plot(X_grid,model.predict(X_grid))
plt.xlabel("grade")
plt.ylabel("Salary")
plt.show()
```
