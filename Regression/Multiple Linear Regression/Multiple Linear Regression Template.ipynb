# Multiple Linear Regression 

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing Data Set

```python
df=pd.read_csv('covidDataSetforMultipleLinearRegression.csv')
X=df.iloc[:,0:4]
y=df.iloc[:,4:5]
```

```python
print(X)
print(y)
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[1])],remainder="passthrough")
X=np.array(ct.fit_transform(X))
```

```python
print(X)
```

## Splitting into test and training set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=0)
print(X_test)
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

## Predicting the Test set results

```python
y_pred=model.predict(X_test)
np.set_printoptions(precision=2)
result=np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.to_numpy().reshape(len(y_test),1)), 1)
```

```python
print(result)
```

## Visualizing results

```python
a=result[:,0]
b=result[:,1]
```

```python
plt.plot(a)
plt.plot(b)
plt.show()
```

## Test the accuracy of the result


```python
from sklearn.metrics import r2_score
accuracy=r2_score(y_test,y_pred)
print("Accuracy of the test is :",accuracy*100)
```
