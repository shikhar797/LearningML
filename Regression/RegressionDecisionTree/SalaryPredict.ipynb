## import all the use full libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

## create data set and variable


```python
df=pd.read_csv("salary.csv")
X=df.iloc[:,1:2].values
y=df.iloc[:,2:].values
```

## train data

```python
model=DecisionTreeRegressor(random_state=42)
model.fit(X,y)
```

## predict the value

```python
model.predict([[3.1]])
```

## visual representation

```python
X_grid = np.arange(X.min(), X.max(), 0.01)
X_grid = X_grid.reshape(-1, 1) #-1 represent that calculate row by yourself or can write reshape(len(X),1)
plt.scatter(X,y)
plt.plot(X_grid,model.predict(X_grid))
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 10))
plot_tree(
    model,
    feature_names=["Grade"],
)
plt.title("Decision Tree Structure")
plt.show()
```
