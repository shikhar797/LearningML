```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('AmazonClustering.csv')
X=df.iloc[:,[2,4]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    model=KMeans(n_clusters=i,random_state=42,init='k-means++')
    model.fit(X)
    wcss.append(model.inertia_)
print(wcss)    
```

```python
plt.scatter(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of cluster")
plt.ylabel("wcss value")
plt.plot(range(1,11),wcss)
plt.show()
```

```python
kmeanModel=KMeans(n_clusters=4,init='k-means++',random_state=42)
y_pred=kmeanModel.fit_predict(X)
```

```python
print(y_pred)
```

```python
plt.scatter(X[y_pred==0,0],X[y_pred==0,1],c='red',label='Cluster 1')
plt.scatter(X[y_pred==1,0],X[y_pred==1,1],c='blue',label='Cluster 2')
plt.scatter(X[y_pred==2,0],X[y_pred==2,1],c='green',label='Cluster 3')
plt.scatter(X[y_pred==3,0],X[y_pred==3,1],c='yellow',label='Cluster 4')
plt.scatter(kmeanModel.cluster_centers_[:,0],kmeanModel.cluster_centers_[:,1],c='black',label='centroid')
plt.show()
```
