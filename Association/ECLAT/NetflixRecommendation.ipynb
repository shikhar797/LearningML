```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Netflix movie recommendation.csv')
```

```python
df
```

```python
trans=[]
for i in range(0,7466):
    l=[]
    for j in range(0,20):
        l.append(str(df.values[i,j]))
    trans.append(l)    
```

```python
from apyori import apriori
result=apriori(transactions=trans,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(result)
print(results)
```

```python
def inspect(results):
    movie1=[list(result.items)[0] for result in results]
    movie2=[list(result.items)[1] for result in results]
    support=[result.support for result in results]
    return list(zip(movie1,movie2,support))
movie_recommendation=pd.DataFrame(inspect(results),columns=['movie1','movie2','Support'])
print(movie_recommendation.nlargest(n=10,columns='Support'))
```
