```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Big_Basket.csv')
```

```python
trans=[]
for i in range(0,7218):
    l=[]
    for j in range(0,20):
        l.append(str(df.values[i,j]))
    trans.append(l)    
print(trans)
# in list comprehension we can write it as 
# for i in range(0,7218):
##     trans.append([df.values[i,j] for  j in range(0,20)])
```

```python
print(np.shape(trans))
```

```python
from apyori import apriori
basket_analysis=apriori(transactions=trans,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

```python
print(basket_analysis)
```

```python
result=list(basket_analysis)
```

```python
def inspect(results):
    product1=[list(result.items)[0] for result in results]
    product2=[list(result.items)[1] for result in results]
    support=[result.support for result in results]
    confidence=[result.ordered_statistics[0][2] for result in results]
    lift=[result.ordered_statistics[0][3] for result in results]
    return list(zip(product1,product2,support,confidence,lift))

market_basket_analysis=pd.DataFrame(inspect(result),columns=['product 1','product 2','support','confidence','lift'])
print(market_basket_analysis.nlargest(n=10,columns='lift'))
```
