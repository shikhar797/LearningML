```python
import pandas as pd
```

```python
df=pd.read_csv('breastcancer.csv')
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
##feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
LogiRegModel=LogisticRegression(random_state=42)
LogiRegModel.fit(X_train,y_train)
LogiReg_y_pred=LogiRegModel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(y_test,LogiReg_y_pred))
print(f"Accuracy of the model is :{accuracy_score(y_test,LogiReg_y_pred)*100}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNNmodel=KNeighborsClassifier(n_neighbors=5)
KNNmodel.fit(X_train,y_train)
knn_y_pred=KNNmodel.predict(X_test)

print(confusion_matrix(y_test,knn_y_pred))
print(accuracy_score(y_test,knn_y_pred)*100)
```

```python
from sklearn.naive_bayes import GaussianNB
naivebayesModel=GaussianNB()
naivebayesModel.fit(X_train,y_train)
naiveBayes_y_pred=naivebayesModel.predict(X_test)

print(confusion_matrix(y_test,naiveBayes_y_pred))
print(accuracy_score(y_test,naiveBayes_y_pred)*100)
```

```python
from sklearn.svm import SVC
SVCmodel=SVC(kernel='rbf',random_state=42)
SVCmodel.fit(X_train,y_train)

SVC_y_pred=SVCmodel.predict(X_test)

print(confusion_matrix(y_test,SVC_y_pred))
print(accuracy_score(y_test,SVC_y_pred)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
DecTreeClass_model=DecisionTreeClassifier(criterion='entropy',random_state=42)
DecTreeClass_model.fit(X_train,y_train)
DecTreeClas_y_pred=DecTreeClass_model.predict(X_test)

print(confusion_matrix(y_test,DecTreeClas_y_pred))
print(accuracy_score(y_test,DecTreeClas_y_pred)*100)
```
