# Simple Linear Regression 

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing Data Set

```python
data_set=pd.read_csv('salary.csv')
X=data_set.iloc[:,0:1].values
y=data_set.iloc[:,1:].values
```

```python
print(X)
```

```python
print(y)
```

## Splitting into test and training set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_test)
```

```python
print(X_train)
```

```python
print(y_test)
```

```python
print(y_train)
```

## Training on Training Set using Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(X_train,y_train)
LinearRegression()
```

## Prediction on Test Set

```python
y_pred=model.predict(X_test)
```

## Visualizing the Training Set

```python
plt.scatter(X_train,y_train,color="red")
plt.plot(X_train,model.predict(X_train),color='red')
plt.title("CTC based on years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("CtC")
plt.show()
```

## Visualizing the Test Set

```python
plt.scatter(X_test,y_test,color='blue')
plt.plot(X_train,model.predict(X_train),color='blue')
plt.title("CTC based on years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("CtC")
plt.show()
```

## Testing the accuracy of Model

```python
from sklearn.metrics import r2_score

y_pred=model.predict(X_test)

accuracy=r2_score(y_test,y_pred)
print("Accuracy of the model is ",int(accuracy*100),"%")
```
