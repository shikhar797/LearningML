### So objective of this file is apply Simple Linear regression on Ice cream revenue based on Temperature


### 1.import necessary python modules

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### 2.Load the Data set

```python
data_set=pd.read_csv("iceCreamRevenue.csv")
X=data_set.iloc[:,0:1].values
y=data_set.iloc[:,1:].values
```

```python
print(X)
```

```python
print(y)
```

### 3.Divide the data set into train test data set

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.3,random_state=42)
```

```python
print(X_train)
```

### 4.Training training data set in Simple Linear Regression

```python
model=LinearRegression()

model.fit(X_train,y_train)
LinearRegression()
```

### 5.Predit the test data output

```python
y_predict=model.predict(X_test)
```

### 6.Visulaize the training data into graph

```python
plt.scatter(X_train,y_train,color="red",label="Actual training data")
plt.plot(X_train,model.predict(X_train),color="blue",label="Regression line")      #draws the line predicted by the model
plt.legend()
plt.title("Ice Cream revenue based on Teperature")
plt.xlabel("Temperature")
plt.ylabel("Revenue")
plt.show()
```

### 7.Visualizing the test data set

```python
plt.scatter(X_test,y_predict,color="red",label="actual test data") 
plt.plot(X_train,model.predict(X_train),color="blue",label="Regression line") #draws the line predicted by the model
plt.legend()
plt.title("Ice Cream revenue based on Teperature")
plt.xlabel("Temperature")
plt.ylabel("Revenue")
plt.show()
```

### 8.Finding the accuracy


```python
from sklearn.metrics import r2_score

y_pred=model.predict(X_test)
accuracy=r2_score(y_test,y_pred)

print("Acuracy of the model is ",accuracy)
```
