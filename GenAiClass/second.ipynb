```python
import nltk
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
reviews=[
    "The movie was fantastic and the actor did a great job.",
    "i did not like the plot , it was too predicatable and boring.",
    "Amazing Cinemography but the storyline was weak and slow.",
    "The movie is good but not exceptionally , could have been better."
]
```

```python
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

```

```python
def preprocess_text(text):
    text = text.lower() # To convert into lowercase characters
    text = text.translate(str.maketrans('', '', string.punctuation)) # To Remove Punctuations
    tokens = nltk.word_tokenize(text) # Tokenizer
    return tokens
```

```python
processed_review=[preprocess_text(review) for review in reviews]
```

```python
stop_words=set(stopwords.words('english'))
```

```python
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]
```

```python
reviews_no_stopwords=[remove_stopwords(token) for token in processed_review]
```

```python
for i ,review in enumerate(reviews):
    print(f"\n Original Review {i+1}:{processed_review[i]}")
    print(f"\n Original Review {i+1}:{reviews_no_stopwords[i]}")
```

```python
orginal_vocab=set([word for tokens in processed_review for word in tokens])
filtered_vocab=set([word for tokens in reviews_no_stopwords for word in tokens])

print(f"Original vocabular size :{len(orginal_vocab)}")
print(f"Vocabularly size after stop word removal :{len(filtered_vocab)}")
```
