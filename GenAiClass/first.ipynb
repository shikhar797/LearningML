```python
from nltk.corpus import brown
```

```python
print(len(brown.categories()))
print(brown.categories())
```

```python
from nltk.tokenize import word_tokenize
categories=brown.categories()
print(f"Number of categories is :{len(categories)}")
```

```python
category_word_count={cat:len(brown.words(categories=cat)) for cat in categories}
```

```python
category_word_count.items()
```

```python
for cat,count in category_word_count.items():
    print(f"{cat}:{count}")
```
