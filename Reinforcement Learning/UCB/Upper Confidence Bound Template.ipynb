# Upper Confidence Bound

## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing dataset

```python
df=pd.read_csv('facebookad.csv')
```

## Implementation

```python
import math
rows=15000
num_ads=10
ad_selected=[]

number_of_selection=[0]*num_ads

sum_of_rewards=[0]*num_ads
total_reward=0

for n in range(0,rows):
    ad=0
    max_upper_confidence_bound=0
    for i in range(0,num_ads):
        if(number_of_selection[i]>0):
            average_reward=sum_of_rewards[i]/number_of_selection[i]
            delta_i=math.sqrt((3/2)*(math.log(n+1)/number_of_selection[i]))
            upper_bound=delta_i+average_reward
        else:
            upper_confidence_bound=1e300
            if upper_confidence_bound>max_upper_confidence_bound:
                max_upper_confidence_bound=upper_confidence_bound
                ad=i
```
