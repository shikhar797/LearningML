```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]

plt.scatter(x1,y1)
plt.show()
```

```python
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]

plt.scatter(x1,y1)
plt.plot(x1,y1)
plt.show()
```

```python
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]

x2=[5,4,3,2,1]
y2=[1,2,3,4,5]
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
x=np.array([1,2,3,4,5,6])
y=2*(x**4)+100
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=2*(x**4)+100
plt.plot(x,y)
plt.show()
```
