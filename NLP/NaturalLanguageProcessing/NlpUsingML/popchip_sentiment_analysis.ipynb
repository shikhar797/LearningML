```python
import pandas as pd
import re 
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
def remove_lower_pun(sentence):
    sentence=sentence.lower()
    sentence=re.sub(r'[^\w\s]', '', sentence)
    return sentence
def token_lemma_stop(sentence):
    doc=nlp(sentence)
    output=[token.lemma_ for token in doc if not token.is_stop]
    return ' '.join(output)
```

```python
df=pd.read_excel('../../../Popchip_Reviews.xlsx')
```

```python
df
```

```python
df['Text']=df['Text'].apply(remove_lower_pun)
df['Text']=df['Text'].apply(token_lemma_stop)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',min_df=0.2)
cv_mat=cv.fit_transform(df['Text'])
cv_mat_df=pd.DataFrame(cv_mat.toarray(),columns=cv.get_feature_names_out())
```

```python
cv_mat_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english',min_df=0.2)
tfidf_mat=tfidf.fit_transform(df['Text'])
tfidf_mat_df=pd.DataFrame(tfidf_mat.toarray(),columns=tfidf.get_feature_names_out())
```

```python
tfidf_mat_df
```

```python
y=df['Priority']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(cv_mat_df,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report
model=MultinomialNB()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)

print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

model1=LogisticRegression()

X1_train,X1_test,y1_train,y1_test=train_test_split(tfidf_mat_df,y,test_size=0.2)

model1.fit(X1_train,y1_train)

y1_pred=model1.predict(X1_test)

print(accuracy_score(y1_test,y1_pred))

print(classification_report(y1_test,y1_pred))


```
