```python
import spacy
nlp=spacy.load('en_core_web_sm')    # you have to download it before using it -> python -m spacy download en_core_web_sm
```

```python
phrase="i am selling lemons for $5"
```

```python
doc=nlp(phrase)
```

```python
doc    #this doc contain tokens
```

```python
[token.text for token in doc]          
```

```python
[token.lemma_ for token in doc]              # this lemma_ function is inside spacy and this if for lemmetization
```

```python
# spacy does not support stemming so we will use nltk for it
#lemmetization is better then stemming for text preprocessing 
new_phrase="I am very happy seeing your happiness"
new_phrase=nlp(new_phrase)
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
[stemmer.stem(word.text) for word in new_phrase]

#stemming is genrally good for speed ,we use in place such as browser searching etc
```

```python
nlp.Defaults.stop_words        # to view some of the stop words
```

```python
# removing top words from the sentence

[token.lemma_ for token in doc if not token.is_stop]
```

```python
[{token.text,token.pos_} for token in doc ]
```
